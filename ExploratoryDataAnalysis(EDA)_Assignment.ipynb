# Step 1: Understanding the Problem and the Data

### Business Understanding & Goal Setting

The purpose of this analysis is to understand the overall performance of the Superstore business by exploring its sales data. The goal is to identify which factors most influence sales and profit, such as region, product category, or discount. By analyzing this dataset, we aim to uncover patterns and trends that can help improve decision-making related to pricing, inventory, and marketing.

Business Goal / Research Question:
Which regions, product categories, and customer segments contribute the most to profit, and how do discounts impact sales and profitability?

Target / Key Variables:

 - Sales (target variable for performance)

 - Profit, Discount, Region, Category, Segment (key drivers)

Expected Outcomes / Valuable Insights:

 - Identify the most and least profitable categories or regions.

 - Understand the relationship between discounts and profit margins.

 - Suggest areas for business improvement (e.g., reducing losses, targeting specific segments).

### Evaluation Criteria (KPI or Metric)

To measure the success of the analysis, we can use:

Profit Margin (%) = (Profit / Sales) × 100
This KPI helps evaluate which regions or products are most efficient in generating profit relative to sales.
A higher profit margin indicates better business performance.

### Data Quality Review

Initial inspection shows that the dataset is mostly clean and well-structured. However, potential issues include:

 - Outliers in Sales and Profit due to unusually large transactions.

 - Some imbalances in category and regional data.

 - Few or no missing values, which simplifies cleaning.

These issues may slightly influence the accuracy of summary statistics or model training later.

### Domain and Ethical Considerations

The dataset contains business transaction data, so there are no major ethical concerns as it doesn’t include personal customer information. However, care should be taken to ensure that business-sensitive data (e.g., internal profit figures) is not misused. From a domain perspective, the analysis should comply with general data privacy and confidentiality practices used in retail analytics.

```python

import missingno as msno 
# -----------------------------------------------------
# 1. Identify Missing Data
# -----------------------------------------------------

missing_counts = df.isnull().sum()
missing_percent = (missing_counts / len(df)) * 100
missing_data = pd.DataFrame({
    'Missing_Count': missing_counts,
    'Missing_%': missing_percent
})
print("=== Missing Data Overview ===")
display(missing_data)

# -----------------------------------------------------
# 2. Visualize Missing Data Pattern
# -----------------------------------------------------

plt.figure(figsize=(8, 5))
sns.heatmap(df.isnull(), cbar=False, cmap='viridis')
plt.title("Missing Data Heatmap", fontsize=14)
plt.show()

# Alternatively, use missingno matrix
msno.matrix(df)
plt.show()

# -----------------------------------------------------
# 3. Diagnose Missingness Mechanism
# -----------------------------------------------------

"""
Missingness Diagnosis
---------------------
Since there are no missing values in this dataset, 
we can reasonably assume the data is 'MCAR' (Missing Completely at Random).
However, if missingness existed, we'd explore whether it was:
    • MCAR - Missing completely at random
    • MAR  - Missing dependent on another variable
    • MNAR - Missing dependent on the missing value itself
"""

# -----------------------------------------------------
# 4. Choose and Justify Handling Method
# -----------------------------------------------------

"""
Since the dataset contains 0 missing values,
no imputation or removal is required.

If missing data existed, the general strategy would be:
    • Numerical columns: impute using mean/median
    • Categorical columns: impute using mode
    • Drop columns with >30% missing data if not critical
"""

# Example of how you *would* handle missing data (for reference):

for col in df.columns:
    if df[col].isnull().sum() > 0:
        if df[col].dtype in ['float64', 'int64']:
            df[col].fillna(df[col].median(), inplace=True)
        else:
            df[col].fillna(df[col].mode()[0], inplace=True)

# -----------------------------------------------------
# 5. Post-Imputation Check
# -----------------------------------------------------

print("=== Post-Imputation Missing Value Check ===")
display(df.isnull().sum().to_frame('Remaining Missing Values'))

# -----------------------------------------------------
# 6. Summary Table
# -----------------------------------------------------

missing_summary = pd.DataFrame({
    'Column': df.columns,
    'Before_Missing': missing_counts.values,
    'After_Missing': df.isnull().sum().values,
    'Action_Taken': ['No Action Needed'] * len(df.columns)
})

print("=== Missing Data Handling Summary ===")
display(missing_summary)

# -----------------------------------------------------
# 7. Reflection
# -----------------------------------------------------

"""
Reflection
----------
Since there are no missing values in the Superstore dataset, 
the data is clean with respect to completeness.
This reduces the risk of bias and simplifies preprocessing.
Future steps will focus on:
    • Handling outliers
    • Encoding categorical variables
    • Feature scaling (if modeling is applied)
"""
```

```python



# Import required libraries
import pandas as pd
import numpy as np
from scipy import stats
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
file_path = r'C:\Users\DELL\OneDrive\Desktop\Saylani-ML\Superstore.csv'
df = pd.read_csv(file_path, encoding='latin1')

# Preview dataset
print("===== Dataset Preview =====")
print(df.head())

# Select numerical columns for analysis
num_cols = ['Sales', 'Quantity', 'Discount', 'Profit']

# ------------------------------------------------------
# 1a. CENTRAL TENDENCY: Mean, Median, Mode
# ------------------------------------------------------

print("\n===== Measures of Central Tendency =====")
for col in num_cols:
    mean_val = df[col].mean()
    median_val = df[col].median()
    mode_val = df[col].mode()[0]
    print(f"\n{col}:")
    print(f"Mean   = {mean_val:.2f}")
    print(f"Median = {median_val:.2f}")
    print(f"Mode   = {mode_val:.2f}")

# ------------------------------------------------------
# 1b. MEASURES OF SPREAD: Std, Variance, Range, IQR
# ------------------------------------------------------

print("\n===== Measures of Spread =====")
for col in num_cols:
    std_val = df[col].std()
    var_val = df[col].var()
    range_val = df[col].max() - df[col].min()
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    print(f"\n{col}:")
    print(f"Standard Deviation = {std_val:.2f}")
    print(f"Variance           = {var_val:.2f}")
    print(f"Range              = {range_val:.2f}")
    print(f"IQR (Q3-Q1)        = {IQR:.2f}")

# ------------------------------------------------------
# 1c. SKEWNESS & KURTOSIS
# ------------------------------------------------------

print("\n===== Skewness and Kurtosis =====")
for col in num_cols:
    skew = df[col].skew()
    kurt = df[col].kurt()
    print(f"\n{col}:")
    print(f"Skewness = {skew:.2f}")
    print(f"Kurtosis = {kurt:.2f}")

# Create a combined summary table
summary = pd.DataFrame({
    'Mean': df[num_cols].mean(),
    'Median': df[num_cols].median(),
    'Mode': [df[c].mode()[0] for c in num_cols],
    'Std Dev': df[num_cols].std(),
    'Variance': df[num_cols].var(),
    'Range': df[num_cols].max() - df[num_cols].min(),
    'IQR': df[num_cols].quantile(0.75) - df[num_cols].quantile(0.25),
    'Skewness': df[num_cols].skew(),
    'Kurtosis': df[num_cols].kurt()
}).round(2)

print("\n===== Summary Table =====")
display(summary)
```

## 3: OUTLIER DETECTION AND HANDLING

```python
# Selected numerical columns
num_cols = ['Sales', 'Quantity', 'Discount', 'Profit']

# ------------------------------------------------------
# 3a. Detect Outliers using Z-score method
# ------------------------------------------------------
print("===== Z-SCORE OUTLIER DETECTION =====")
z_scores = np.abs(stats.zscore(df[num_cols]))
outliers_z = (z_scores > 3)  # Common threshold
outlier_counts_z = outliers_z.sum()
print(outlier_counts_z)

# ------------------------------------------------------
# 3b. Detect Outliers using IQR method
# ------------------------------------------------------
print("\n===== IQR OUTLIER DETECTION =====")
outlier_counts_iqr = {}

for col in num_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    outliers = df[(df[col] < lower) | (df[col] > upper)]
    outlier_counts_iqr[col] = len(outliers)

iqr_summary = pd.DataFrame.from_dict(outlier_counts_iqr, orient='index', columns=['Outlier Count'])
display(iqr_summary)

# ------------------------------------------------------
# 3c. Visualize Outliers using Box Plots (Before Removal)
# ------------------------------------------------------
plt.figure(figsize=(12, 6))
for i, col in enumerate(num_cols, 1):
    plt.subplot(2, 2, i)
    sns.boxplot(x=df[col], color='lightcoral')
    plt.title(f'Before Removal: {col}', fontsize=12)
plt.tight_layout()
plt.show()

# ------------------------------------------------------
# 3d. Handle Outliers — Using IQR method (Remove)
# ------------------------------------------------------
df_clean = df.copy()

for col in num_cols:
    Q1 = df_clean[col].quantile(0.25)
    Q3 = df_clean[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    df_clean = df_clean[(df_clean[col] >= lower) & (df_clean[col] <= upper)]

print(f"\nOriginal dataset shape: {df.shape}")
print(f"Cleaned dataset shape: {df_clean.shape}")

# ------------------------------------------------------
# 3e. Visualize Outliers After Removal
# ------------------------------------------------------
plt.figure(figsize=(12, 6))
for i, col in enumerate(num_cols, 1):
    plt.subplot(2, 2, i)
    sns.boxplot(x=df_clean[col], color='lightgreen')
    plt.title(f'After Removal: {col}', fontsize=12)
plt.tight_layout()
plt.show()
```

# Step 6:Handling Outliers

```python
#1 
import numpy as np

# Select numeric columns
numeric_cols = ['Sales', 'Profit', 'Discount', 'Quantity']

# Detect outliers using IQR
for col in numeric_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_limit = Q1 - 1.5 * IQR
    upper_limit = Q3 + 1.5 * IQR

    outliers = df[(df[col] < lower_limit) | (df[col] > upper_limit)]
    print(f"{col}: {len(outliers)} outliers detected")

    # Optional visualization
    plt.figure(figsize=(6, 4))
    sns.boxplot(x=df[col], color='lightblue')
    plt.title(f"Boxplot of {col} (Outlier Detection)")
    plt.show()
```

```python
 #2
# Winsorize (Cap) outliers
for col in ['Sales', 'Profit']:
    lower_cap = df[col].quantile(0.01)
    upper_cap = df[col].quantile(0.99)
    df[col] = np.where(df[col] < lower_cap, lower_cap, df[col])
    df[col] = np.where(df[col] > upper_cap, upper_cap, df[col])

print("Outliers capped successfully.")
```

### 3. Explaining the Impact of Outliers:

Outliers were detected mainly in the Sales and Profit columns, caused by a few large transactions. Completely removing these points would have led to the loss of valuable business insights, so a Winsorization approach was applied, capping values at the 1st and 99th percentiles. This reduced their influence on averages and correlations without discarding valid data. As a result, statistical measures and visualizations are now more balanced, improving the reliability of future modeling steps.

### 4. Justification and Conclusion:
The IQR method was used to detect outliers, as it’s best suited for the skewed distribution of financial data in this dataset. Outliers were found mostly in Sales and Profit, representing genuine high-value orders rather than errors. Therefore, they were capped instead of removed, ensuring data integrity while reducing the impact of extreme values. Handling outliers in this way enhances the accuracy of analysis and ensures that subsequent models are not overly influenced by rare, extreme data points.
